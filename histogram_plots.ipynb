```python
import glob
import numpy as np
import rasterio
import matplotlib.pyplot as plt
from datetime import datetime
from collections import defaultdict

base_path = "/Users/sameeps/Documents/Dr-Malof/WIldfire/WildfireSpreadTS/Data/WildfireSpreadTsData"
years = [2018, 2019, 2020, 2021]

for year in years:
    print(f"\nProcessing year {year}...")

    # Use a defaultdict to accumulate sums per day
    day_sum_intensities = defaultdict(float)

    tifs = glob.glob(f"{base_path}/{year}/*/*.tif")
    tifs.sort()

    for tif_file in tifs:
        # parse DOY
        date_str = tif_file.split("/")[-1].replace(".tif", "")
        doy = datetime.strptime(date_str, "%Y-%m-%d").timetuple().tm_yday

        with rasterio.open(tif_file) as src:
            if src.count >= 23:
                fire_data = src.read(23)
                fire_mask = fire_data > 0
            else:
                continue

        sum_intensity = fire_data[fire_mask].sum() if np.any(fire_mask) else 0

        # Accumulate total sum for this day across all fires
        day_sum_intensities[doy] += sum_intensity

    # Prepare for plotting
    days = sorted(day_sum_intensities.keys())
    total_sum_per_day = [day_sum_intensities[day] for day in days]

    # DEBUG print
    for day, total in zip(days, total_sum_per_day):
        print(f"DOY {day}: total sum of band 23 values = {total}")

    # Plot DOY vs total sum
    plt.figure(figsize=(10,5))
    plt.bar(days, total_sum_per_day, color='firebrick', width=1.0)
    plt.title(f"Total sum of band 23 values (all fires) by day in {year}")
    plt.xlabel("Day of Year (1-365)")
    plt.ylabel("Total sum of fire band values")
    plt.grid(True)
    plt.show()
```
